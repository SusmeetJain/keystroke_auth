```python
from tensorflow.keras.models import model_from_json

def ex_model(data, user_name):
    x1 = np.array(data.loc[data.subject == user_name, 'H.period':'H.Return'][35:36]) #subj is user_name
    x2 = np.array(data.iloc[-1:,3:])
    json_file = open('model_architecture_new.json', 'r')
    loaded_model_json = json_file.read()
    json_file.close()
    loaded_model = model_from_json(loaded_model_json)
    # load weights into new model
    loaded_model.load_weights("model_weights_new.h5")
    print("Loaded model from disk")
    loaded_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    score = np.rint(loaded_model.predict([x1, x2]))
    return int(score[0][0])
```
